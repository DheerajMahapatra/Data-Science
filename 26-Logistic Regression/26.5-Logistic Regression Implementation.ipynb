## Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=15)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# Model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

## Hyperparameter Tuning And Cross Validation

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
dict(penalty=penalty,C=c_values,solver=solver)
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Randomized SearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
y_pred=randomcv.predict(X_test)
```

## Logistic Regression For Multiclass Classification

```python
## Create the Dataset
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=15)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

### New version

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

logistic = OneVsRestClassifier(LogisticRegression())

logistic.fit(X_train, y_train)

y_pred = logistic.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Hyperparameter Tuning And Cross Validation

```python
ovr_model = OneVsRestClassifier(
    LogisticRegression(max_iter=1000)
)
```

```python
# params = {
#     'estimator__penalty': ['l2', None],
#     'estimator__C': [100, 10, 1.0, 0.1, 0.01],
#     'estimator__solver': ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag']
# }
params = {
    'estimator__C': [100,10,1,0.1,0.01],
    'estimator__penalty': ['l1','l2','elasticnet'],
    'estimator__solver': ['saga'],
    'estimator__l1_ratio': [0.2,0.5,0.8]
}
```

```python
grid = GridSearchCV(
    estimator=ovr_model,
    param_grid=params,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
```

```python
grid.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Randomized SearchCV

```python
model = LogisticRegression(max_iter=1000)
```

```python
param_dist = {
    'C': np.logspace(-4, 4, 20),
    'penalty': ['l2'],
    'solver': ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'class_weight': [None, 'balanced']
}
```

```python
randomcv = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=20,
    scoring='accuracy',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
randomcv.fit(X_train, y_train)
```

```python
print("Best Parameters:", randomcv.best_params_)
print("Best Score:", randomcv.best_score_)
```

```python
best_model = randomcv.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
