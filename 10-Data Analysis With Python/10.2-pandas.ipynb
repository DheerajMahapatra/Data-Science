## Numpy
#### NumPy is a fundamental library for scientific computing in Python. It provides support for arrays and matrices, along with a collection of mathematical functions to operate on these data structures. In this lesson, we will cover the basics of NumPy, focusing on arrays and vectorized operations.

### NumPy is used for fast numerical computations using arrays, while Pandas is used for data manipulation and analysis. Pandas is built on top of NumPy and internally uses NumPy arrays for efficient data handling.

```python
# Series
### A Pandas Series is a one-dimensional array-like object that can hold any data type. it is similar to a column in a table.

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## Create a Series from dictionary

data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10, 20, 30]
index = ['a','b','c']
pd.Series(data, index = index)
```

```python
## Dataframe
## create a Dataframe from a dictionary of list

data={
    'Name':['Dheeraj','John','Jack'],
    'Age':[21,30,45],
    'City':['Bangalore','New York','Florida']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Dheeraj', 'Age':21, 'City':'Bangalore'},
    {'Name':'John', 'Age':34, 'City':'Bangalore'},
    {'Name':'Bappy', 'Age':32, 'City':'Bangalore'},
    {'Name':'Jack', 'Age':32, 'City':'Bangalore'}
]

df=pd.DataFrame(data)
print(df)
print(type(df))
```

| Feature    | loc                   | iloc                   |
| ---------- | --------------------- | ---------------------- |
| Index type | label                 | integer position       |
| Syntax     | `df.loc[row, column]` | `df.iloc[row, column]` |
| End index  | included              | excluded               |
| Use        | names / labels        | numeric position       |

| Code                         | Kya hota hai              |
| ---------------------------- | ------------------------- |
| `df.drop(... )`              | new DataFrame return      |
| `df.drop(..., inplace=True)` | original DataFrame modify |

##### When inplace=False (default), the function returns a new DataFrame which Jupyter displays automatically. When inplace=True, the original DataFrame is modified and the function returns None, so nothing is displayed.
