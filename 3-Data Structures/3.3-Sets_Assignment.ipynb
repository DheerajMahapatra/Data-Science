### Assignment 3: Set Operations

Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.



```python
first_five = set(range(1,6))
print(f"First five positive(+) integers are : \n {first_five} \n")

even_five = set(range(2,11,2))
print(f"First five even integers are : \n {even_five} \n")

# Union
union_sets = first_five.union(even_five)
print(f"Union of the sets are : \n {union_sets} \n")

# Intersection
intersection_sets = first_five.intersection(even_five)
print(f"Intersection of the sets are : \n {intersection_sets} \n")

# Difference
difference_sets = first_five.difference(even_five)
print(f"Difference of the sets are : \n {difference_sets} \n")

# Symmetric Difference
symmetric_difference_sets = first_five.symmetric_difference(even_five)
print(f"Symmetric Difference of the sets are : \n {symmetric_difference_sets}")
```

### Assignment 7: Subsets and Supersets

Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. Check if the second set is a subset of the first set and if the first set is a superset of the second set. Print the results.



```python
first_five = set(range(1,6))
print(f"First five positive integers are : \n {first_five} \n")

first_three = set(range(1,4))
print(f"First three positive integers are : \n {first_three} \n")

subset = first_three.issubset(first_five)
superset = first_five.issuperset(first_three)

print(f"Is Set 2 is a subset of set 1 ? \n Ans:-{first_three.issubset(first_five)} \n")
print(f"Is Set 1 is a superset of set 2 ? \n Ans:-{first_five.issuperset(first_three)} \n")

if (subset == True and superset == True):
    print("Yes! \n Set 2 is a subset of set 1 AND Set 1 is a superset of set 2.")
    
else:
    print("No! \n Set 2 is NOT a subset of set 1 AND Set 1 is NOT a superset of set 2.")
```
