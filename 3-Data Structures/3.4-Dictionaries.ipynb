### Dictionaries

```python
## Creating Dictionaries

empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Dheeraj", "age":21,"grade":25}
print(student)
print(type(student))
```

```python
# Single key is always used

student = {"name":"Dheeraj", "age":21,"name":25}
print(student)
```

```python
## Accessing Dictionary Elements

student = {"name":"Dheeraj", "age":21,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary Elements

print(student['grade'])
print(student['age'])


## Accessing using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', "Not Available"))
```

```python
## Modifying DIctionary Elements

print(student )
```

```python
student["age"] = 29     #Update value for the key
print(student)
student["address"] = "India"    #Added a new key and value
print(student)
```

```python
del student['grade']    #Delete key and value pair
print(student)
```

```python
## Dictionary methods

keys = student.keys()   #get all the key
print(keys)
value = student.values()    #got all values
print(value)

items = student.items()     #get all key value pairs
print(items)
```

```python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Krish"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() #shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "Rupu"
print(student_copy1)
print(student)
```

```python
## Iterating Over Dictionaries

# Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries

students = {
    "student1":{"name":"Dheertaj","age":21},
    "student2":{"name":"Rupu","age":18}
}
print(students)
```

```python
## Access nested dictionaries elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension

squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension

evens= {x:x**2 for x in range(10) if x%2 == 0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
#frequency = set(numbers)
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merged 2 dictionary into one

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
