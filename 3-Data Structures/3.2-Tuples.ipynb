```python
## Immutable Nature of Tuples

lst = [1,2,3,4,5]
print(lst)

lst [1] = "Krish"
print(lst)
```

```python
numbers [1] = "Dheeraj"
```

```python
## Tuple Methods

print(numbers.count(1))
print(numbers. index(3))
```

```python
## Packing and Unpacking Tuple

# Packing Tuple

packed_tuple = 1,"Hello", 3.14
print(packed_tuple)
```

```python
## Unpacking Tuple

a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *

numbers = (1,2,3,4,5,6)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List

lst = [[1,2,3,4],[6,7,8,9],[1, "Hello", 3.14, "c"]]
print(lst[0])
```

```python
lst[0][2]
```

```python
lst[0][0:3]
```

```python
lst = [[1,2,3,4],[6,7,8,9],(1, "Hello", 3.14, "c")]
print(lst[2][0:3])
```

```python
nested_tuple = ((1,2,3), ("a","b","c"), (True, False))

## Access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
