```python
## Accessing List Elements

fruits = ["apple", "banana", "cherry", "kiwi", "guava"]
```

```python
print(fruits[0])
print(fruits[2])
print(fruits[4])
print(fruits[-1])
```

```python
print(fruits[1:])
print(fruits[1:3])
print(fruits[-1:])

print(fruits[-1:-3])
print(fruits[-3:-1])
print(fruits[::-1])
```

```python
## Modify The List Elements
fruits
```

```python
fruits[1] = "watermelon"
print(fruits)
```

```python
fruits[1:] = "watermelon"
```
